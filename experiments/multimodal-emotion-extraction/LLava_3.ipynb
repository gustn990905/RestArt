```python
from lmdeploy import pipeline, ChatTemplateConfig
from lmdeploy.vl import load_image
import asyncio
pipe = pipeline('xtuner/llava-llama-3-8b-hf',
                chat_template_config=ChatTemplateConfig(model_name='llama3'))


image = load_image('white.jpg')

async def main():
  result = pipe(('Please analyze the image above and derive one emotion that is most similar to (happiness, joy, love, sophistication, sensibility, curiosity, awe, sadness, dislike, worry, fear, fear, anger, greed, and sympathy)', image))
  print(result)

loop = asyncio.get_event_loop()
loop.run_until_complete(main())
```
